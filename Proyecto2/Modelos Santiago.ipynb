```
import streamlit
import pandas as pd 
import numpy as np
import seaborn as sns 
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```
df_t=pd.read_csv('logs_modelos_tabasco.csv')
df_t
```

# Regresión Lineal

```
X=df_t.drop(columns=['Poblacion de 0 a 5'])
y=df_t['Poblacion de 0 a 5']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=30,shuffle=True)
```

```
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```
y_pred=pd.Series(reg.predict(X_test))
y_pred.index=y_test.index
Resultados=pd.DataFrame([y_pred,y_test]).T
Resultados.columns=['y_test','y_pred']
Resultados
```

```
mean_squared_error(y_test,y_pred)
```

```
r2_score(y_test,y_pred)
```

# Ridge

```
reg_ridge=Ridge()
reg_ridge.fit(X_train,y_train)
```

```
y_pred=pd.Series(reg_ridge.predict(X_test))
y_pred.index=y_test.index
Resultados=pd.DataFrame([y_pred,y_test]).T
Resultados.columns=['y_test','y_pred']
Resultados
```

```
print('Error Cuadrático Medio:',mean_squared_error(y_test,y_pred))
print('R2:',r2_score(y_test,y_pred))
```

# Lasso

```
reg_lasso=Lasso(alpha=0.05)
reg_lasso.fit(X_train,y_train)
```

```
y_pred=pd.Series(reg_lasso.predict(X_test))
y_pred.index=y_test.index
Resultados=pd.DataFrame([y_pred,y_test]).T
Resultados.columns=['y_test','y_pred']
Resultados
```

```
print('Error Cuadrático Medio:',mean_squared_error(y_test,y_pred))
print('R2:',r2_score(y_test,y_pred))
```

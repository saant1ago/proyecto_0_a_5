## Preparar dataset para los modelos

```
df = df_muestra
X = df.drop('Poblacion de 0 a 5', axis=1)  # Vector de caracterÃ­sticas
y = df['Poblacion de 0 a 5']               # Variable objetivo
```

## Modelo Ãrboles de DecisiÃ³n

```
#Modelo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.23, random_state=23)
model = DecisionTreeRegressor(max_depth = 12, min_samples_split = 7) #HiperparÃ¡metros
model.fit(X_train, y_train)
```

```
#Evaluar modelo
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)
Resultados=pd.DataFrame([y_pred,y_test]).T
Resultados.columns=['y_test','y_pred']
display(Resultados)
print("Error medio cuadrado: ", mse)
print("R^2: ", r2)
```

## Modelo Random Forest

```
# Modelo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.23, random_state=23)
model = RandomForestRegressor(n_estimators = 50, max_depth = 7) #HiperparÃ¡metros
model.fit(X_train, y_train)
```

```
#Evaluar modelo
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)
Resultados=pd.DataFrame([y_pred,y_test]).T
Resultados.columns=['y_test','y_pred']
display(Resultados)
print("Error medio cuadrado: ", mse)
print("R^2: ", r2)
```

## RegresiÃ³n Lineal

```
#Modelo
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.23,random_state=23)
reg=LinearRegression() #No tiene hiperparÃ¡metros considerables para cambiar
reg.fit(X_train,y_train)
```

```
#Evaluar modelo
y_pred = reg.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)
Resultados=pd.DataFrame([y_pred,y_test]).T
Resultados.columns=['y_test','y_pred']
display(Resultados)
print("Error medio cuadrado: ", mse)
print("R^2: ", r2)
```

## Ridge

```
#Modelo
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.23,random_state=23)
reg_ridge=Ridge() #No hay hiperparÃ¡metros considerables para cambiar
reg_ridge.fit(X_train,y_train)
```

```
#Evaluar modelo
y_pred = reg_ridge.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)
Resultados=pd.DataFrame([y_pred,y_test]).T
Resultados.columns=['y_test','y_pred']
display(Resultados)
print("Error medio cuadrado: ", mse)
print("R^2: ", r2)
```

## Lasso

```
#Modelo
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.23,random_state=23)
reg_lasso=Lasso(alpha=0.23) #El Ãºnico hiperparÃ¡metro considerable es alpha
reg_lasso.fit(X_train,y_train)
```

```
#Evaluar modelo
y_pred = reg_lasso.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)
Resultados=pd.DataFrame([y_pred,y_test]).T
Resultados.columns=['y_test','y_pred']
display(Resultados)
print("Error medio cuadrado: ", mse)
print("R^2: ", r2)
```
